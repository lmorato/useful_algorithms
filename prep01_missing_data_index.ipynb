```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"values": [1,0,2,2,0,0,3,3,3,0,0,0,4,4,4,4,0,0,0,0,5,5,5,5,5,0,0,0,0,0,6,6,6,6,6,6,0,0,0,0,0,0]})

print(df)
```

```python
# insert column 'idx' with values of index of df as first column:
df.insert(0, 'idx', df.index)

# create column 'int_bool' integer version of boolean version of df:
df['int_bool'] = df['values'].astype(bool).astype(int)

print(df)
```

```python
# create col neg_pos with -1 in place of 0 in col 'int_bool':
df['neg_pos'] = df['int_bool'].replace(0, -1)

print(df)
```

```python
# delete 'neg_diff_neg_pos' col:
df.drop('neg_diff_neg_pos', axis=1, inplace=True)
```

```python
# create col 'negdiff_negpos' by shifting back col 'neg_pos' by 1 row:
df['negdiff_negpos'] = df['neg_pos'].shift(-1)

print(df)
```
